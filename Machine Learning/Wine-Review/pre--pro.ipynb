```python
import pandas as pd
```

```python
import numpy as np
from geopy.geocoders import Nominatim
import geopy
import numpy
```

```python
nom = Nominatim()
```

```python
df = pd.read_csv("C:/Users/Grand/Desktop/DS-Academy/finalresult.csv")
```

```python
df=df.drop(columns=['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1','region_1','region_2','winery'])
```

```python
df1 = df[df.country=="France"]
```

```python
df1["Address"] = df1['province']+", "+df1['country']
```

```python
hiho= df1['Address'].unique()
```

```python
df2 = pd.DataFrame(hiho,columns=['AddressGeoCode'])
```

```python
df2
```

```python
df2['Address_Geo_Code'] = df2['AddressGeoCode'].apply(nom.geocode)
```

```python
df2['Address_Geo_Code'][0]
```

```python
df2["latitude"]=df2["Address_Geo_Code"].apply(lambda x: x.latitude if x != None else None)
```

```python
df2["longitude"]=df2["Address_Geo_Code"].apply(lambda x: x.longitude if x != None else None)
```

```python
df2['Address']=df2['AddressGeoCode']
```

```python
df3=pd.merge(df1, df2, on='Address')
```

```python
df3
```

```python
df3.dropna(subset=['price','latitude'],axis=0, inplace=True)
```

```python
df3
```

```python
df3=df3.drop_duplicates()
```

```python
df3.to_csv('DataClean&Clear.csv',index=False)
```
