# With PCA

```python
cov_matrix = np.cov(X_train.T)
eig_vals, eig_vects=np.linalg.eig(cov_matrix)
eig_pairs=list(zip(eig_vals,eig_vects.T))
eig_pairs.sort(reverse=True)
sorted_eigVals=[eig_pairs[ind][0] for ind in range(len(eig_pairs))]
sorted_eigVects=[eig_pairs[ind][1] for ind in range(len(eig_pairs))]
var_exp=[var/sum(sorted_eigVals) for var in sorted_eigVals]
var_cum=np.cumsum(sorted_eigVals)/sum(sorted_eigVals)
var_cum
```

```python
var_exp
```

```python
plt.figure(figsize=(8,8))
plt.step(range(1,len(sorted_eigVals)+1), var_cum, where='mid', label='cumulative explained variance', color='k')
plt.bar(range(1,len(sorted_eigVals)+1), var_exp, label='individual explained variance', color='r')
plt.xticks(range(1,len(sorted_eigVals)+1))
plt.legend()
plt.xlabel("Principal Components")
plt.ylabel("Cumulative proportion of variance retained")
plt.title("Retained variance")
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=4)
pca_model=pca.fit_transform(X_train)
```

```python
#The amount of variance that each PC explains
var= pca.explained_variance_ratio_

#Cumulative Variance explains
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
plt.plot(var1)
```

```python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c = kmeans.labels_)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, scaled_df.columns[i], color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.figure(figsize=(10,10))
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel("PC{}".format(1))
plt.ylabel("PC{}".format(2))
plt.grid()

#Call the function. Use only the 2 PCs.
myplot(pca_model[:,0:2],np.transpose(pca.components_[0:2, :]))
plt.show()
```

## Hybrid clustering

```python
kmeansh=KMeans(n_clusters=200)
kmeansh.fit(pca_model)
```

```python
# Classification hierarchique
# Calcul des distances
from scipy import cluster
Z = cluster.hierarchy.ward(kmeansh.cluster_centers_)
# Dendrogram
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(5, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster size')
plt.ylabel('Distance')
dendrogram(
   Z,
   truncate_mode='lastp',  # show only the last p merged clusters
   p=20,  # show only the last p merged clusters
   leaf_rotation=90.,
   leaf_font_size=12.,
   show_contracted=True,  # to get a distribution impression in truncated branches
)
plt.show()
```

```python
k=6
clusters1 = cluster.hierarchy.fcluster(Z, k, criterion='maxclust')
```

```python
clusters1
```

```python
lab=np.empty((len(kmeansh.labels_)))
for i in range(len(clusters1)):
    lab[np.where(kmeansh.labels_==i)[0]]=clusters1[i]
```

```python
plt.scatter(pca_model[:,0], pca_model[:,1], c=lab, cmap='viridis')
```
