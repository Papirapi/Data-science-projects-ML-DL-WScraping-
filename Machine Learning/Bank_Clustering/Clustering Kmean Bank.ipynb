```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df1)
```

```python
df1_2d = pca.transform(df1)
```

```python
ex_df1_2d = pd.DataFrame(df1_2d)
ex_df1_2d.index = df1.index
ex_df1_2d.columns = ['PC1','PC2']
ex_df1_2d.head()
```

```python
print(pca.explained_variance_ratio_)
```

```python
ex_df1_2d.head()
```

```python
#We see that the first PC already explains almost 58% of the variance,
#while the second one accounts for another 21% for a total of almost 79% between the two of them.
```

```python
from sklearn.preprocessing import normalize
    
ex_df1_2d['CLT_mean'] = pd.Series(df1.mean(axis=1), index=ex_df1_2d.index)
CLT_mean_max = ex_df1_2d['CLT_mean'].max()
CLT_mean_min = ex_df1_2d['CLT_mean'].min()
CLT_mean_scaled = (ex_df1_2d.CLT_mean-CLT_mean_min) / CLT_mean_max
ex_df1_2d['CLT_mean_scaled'] = pd.Series(CLT_mean_scaled,index=ex_df1_2d.index) 
ex_df1_2d.head()
```

```python
ex_df1_2d.plot(
    kind='scatter', 
    x='PC2', 
    y='PC1', 
    s=ex_df1_2d['CLT_mean_scaled']*100, 
    figsize=(16,8))
```

```python
ex_df1_2d['CLT_sum'] = pd.Series(df1.sum(axis=1),index=ex_df1_2d.index)
CLT_sum_max = ex_df1_2d['CLT_sum'].max()
CLT_sum_min = ex_df1_2d['CLT_sum'].min()
CLT_sum_scaled =(ex_df1_2d.CLT_sum-CLT_sum_min) / CLT_sum_max
ex_df1_2d['CLT_sum_scaled'] = pd.Series(CLT_sum_scaled,index=ex_df1_2d.index)
ex_df1_2d.plot(kind='scatter',x='PC2', y='PC1',s=ex_df1_2d['CLT_sum_scaled']*100,figsize=(16,8))
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
for k in range(1, 7):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(df1)
    inertia.append(np.sqrt(kmeans.inertia_))
```

```python
plt.plot(range(1, 7), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```
